```python
# PLOT: Radar plot
# You need to define the ranges for each parameter, looking at the chain plots can be useful
ranges = [(1200,2000),
          (2.5,4.7),
          (-0.1,0.8),
          (0,1),
          (1,5),
          (-30,150),
          (-4,-1)]
plotForMoSA.plot_radar(ranges,label='AB Pic b, K band',quantiles=[0.16, 0.5, 0.84],chiffres=[0,2,2,2])
plt.show()
```
